```python
from pathlib import Path
import numpy as np
import pandas as pd
```

```python
CLASS_TO_IDX = {
    "Center": 0,
    "Donut": 1,
    "Edge-Loc": 2,
    "Edge-Ring": 3,
    "Loc": 4,
    "Random": 5,
    "Scratch": 6,
    "Near-full": 7,
    "none": 8,
}

def parse_nested_string(value):
    """Unwrap nested list/ndarray labels into plain string."""
    while isinstance(value, (list, tuple, np.ndarray)):
        if len(value) == 0:
            return  np.nan
        value = value[0]
    s = str(value).strip()
    return s
```

```python
# data paths
input_path = Path('./data/wm811k/LSWMD.pkl')
output_dir = Path('./data/wm811k/preprocessing')
output_dir.mkdir(parents=True, exist_ok=True)

labeled_path = output_dir / 'labeled.pkl'
unlabeled_path = output_dir / 'unlabeled.pkl'
# summary_path = output_dir / 'summary.json'

print('input path:', input_path)
print('output_dir path: ', output_dir)
```

```python
df=pd.read_pickle(input_path)
```

```python
df
```

```python
required_cols = ['waferMap', 'dieSize', 'failureType']
keep_df = df[required_cols].copy()
keep_df['failureType'] = keep_df['failureType'].apply(parse_nested_string)
keep_df['label'] = keep_df['label'] = keep_df["failureType"].map(CLASS_TO_IDX)
keep_df
```

```python
# label data
unlabeled_df = keep_df[keep_df['failureType'].isna()].reset_index(drop=True)
unlabeled_df
```

```python
# label data
labeled_df = keep_df[keep_df['failureType'].notna()].reset_index(drop=True)
labeled_df
```

```python
labeled_df.to_pickle(labeled_path)
unlabeled_df.to_pickle(unlabeled_path)
```
